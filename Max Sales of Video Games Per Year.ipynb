```python
import pandas as pd
import seaborn as sns
import os
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\\Users\sheej.DESKTOP-SDF1007\OneDrive\Documents\data\Cleaned_Video_Game_Sales.csv")
df.head()
```

```python
df_grouped = df.groupby('Platform').sum()[['NA_Sales','EU_Sales','JP_Sales','Other_Sales', 'Global_Sales']]
df_grouped
```

```python
release_sales=df.groupby("Year_of_Release")["Global_Sales","Name"].max() # max sales in each year
release_sales
```

```python
sns.set(style="whitegrid")
f,ax=plt.subplots(figsize=(16,9))
sns.pointplot(x=release_sales.index,y=release_sales.Global_Sales,alpha=0.5,color="red")
plt.text(0,88,'Sales(millions)',fontsize = 17,style = 'italic')
plt.xlabel('Year of Release',fontsize = 15)
plt.ylabel('Max Sales',fontsize = 15)
plt.title('Max Sales of Video Games Per Year',fontsize = 20)
plt.xticks(rotation= 45)
plt.show()
f.savefig('Video_Game_Max_Sales_by_Year.png', dpi=250)
```
