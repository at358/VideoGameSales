```python
import csv
import os
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\sheej.DESKTOP-SDF1007\OneDrive\Documents\data\Video_Games_Sales.csv")
df.head()
```

```python
clean_df = df.dropna(subset = ['Name', 'Platform', 'Year_of_Release', 'Publisher','NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales','Critic_Score','Critic_Count','User_Score','User_Count','Developer','Rating'])
clean_df = clean_df.loc[(clean_df['Name'] != "NaN") & (clean_df['Platform'] != "NaN") & (clean_df['Year_of_Release'] > 0) & (clean_df['Publisher'] != "NaN") &  (clean_df['NA_Sales'] > 0) & (clean_df['EU_Sales'] > 0) & (clean_df['JP_Sales'] > 0) & (clean_df['Other_Sales'] > 0) & (clean_df['Global_Sales'] > 0) & (clean_df['Critic_Score'] > 0) & (clean_df['Critic_Count'] > 0) & (clean_df['User_Score'] > 0) & (clean_df['User_Count'] > 0) & (clean_df['Developer'] != "NaN") & (clean_df['Rating'] != "NaN")]
clean_df
```

```python
clean_df_nd = clean_df.drop_duplicates(subset = ['Name', 'Platform', 'Year_of_Release', 'Publisher'])
clean_df_nd
```

```python
clean_df_nd.to_csv("C:\\Users\sheej.DESKTOP-SDF1007\OneDrive\Documents\data\Cleaned_Video_Game_Sales.csv")
```
